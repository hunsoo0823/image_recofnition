#2. Dictionary

```python
stationfares = []
```

```python
dic = {}
station = input('Enter station: ')
fare = int(input('Enter fare : '))
dic['station'] = station
dic['fare'] = fare
stationfares.append(dic)
print(dic)
print(stationfares)
```

```python
dic = {}
station = '\uC131\uBD81' #@param {type:"string"}
fare =  2000 #@param {type:"integer"}
dic['station'] = station
dic['fare'] = fare
stationfares.append(dic)
print(dic)
print(stationfares)
```
