#2. Loop

```python
num = int(input('Multiplication table of : '))
```

```python
for i in range(1, 10):
    print(num, 'X', i, '=', num*i)
```

```python
for i in range(2, 10):
    for j in range(1, 10):
        print("%2d X %2d = %5d" %(i, j, i*j))
    print("----------------")
```

```python
num = int(input('Multiplication table of : '))
i = 1
while i < 10:
    print(num, 'X', i, '=', num*i)
    i += 1
```

```python
i = 2
while i < 10:
    j = 1
    while j < 10:
        print("%2d X %2d = %5d" %(i, j, i*j))
        j += 1
    print("----------------")
    i += 1
```

```python
stationfares = [
                {"station" : "쳥랑리", "fare" : 1500},
                {"station" : "성북", "fare" : 1800},
                {"station" : "의정부", "fare" : 2000},
                {"station" : "소요산", "fare" : 2500}
]   
```

```python
stations = []
fares = []
for stationfare in stationfares:
    stations.append(stationfare["station"])
    fares.append(stationfare["fare"])
```

```python
print(stations)
print(fares)
```
