#1. Scraping

```python
import requests
url = "http://www.kric.go.kr/jsp/industry/rss/citystapassList.jsp?q_org_cd=A010010021&q_fdate=2020"
```

```python
html_text = requests.get(url).text
```

```python
print(html_text)
```

#2. Parsing

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(html_text, "html.parser")
```

```python
print(soup)
```

#3. Manipulation

```python
tab = soup.find("table", {"class":"listtbl_c100"})
```

```python
print(tab)
```

```python
trs = tab.find('tbody').find_all('tr')
```

```python
trs[1]
```

```python
tdcols = trs[1].find("td", {"class":"tdcol"})
```

```python
print(tdcols)
```

```python
tds =trs[1].find_all('td')
```

```python
print(tds)
```

```python
tds[0]
```

```python
tds[0].text
```

```python
tds[2].text
```

```python
tds[3].text
```

```python
stationpassengers = []
for tr in trs[1:]:
    dic = {}
    tds = tr.find_all('td')
    dic['station'] = tds[0].text
    dic['ride'] = tds[2].text
    dic['alight'] = tds[3].text
    stationpassengers.append(dic)
```
