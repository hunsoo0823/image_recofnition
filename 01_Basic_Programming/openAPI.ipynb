# 1. Open API
 실시간 영업소간 통행시간
 
http://data.ex.co.kr/openapi/trtm/realUnitTrtm


```python
import requests
```

```python
key = '2103045469'
type = 'json'
startUnitCode = '101'
EndUnitCode = '103'

URL = 'http://data.ex.co.kr/openapi/trtm/realUnitTrtm?key='
url = URL + key + '&type=json&iStartUnitCode=' + startUnitCode + '&iEndUnitCode=' + EndUnitCode
```

```python
response = requests.get(url)
```

```python
print(response)
```

```python
json = response.json()
```

```python
json
```

```python
cars = json['realUnitTrtmVO']
```

```python
cars
```

```python
records = []
for car in cars:
    dic = {}
    dic['date'] = car['stdDate']
    dic['time'] = car['stdTime']
    dic['type'] = car['tcsCarTypeDivName']
    records.append(dic)
records
```
