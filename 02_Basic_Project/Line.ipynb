#3. Load Data

#4. 집계시

```python
data_06_22 = data[data.집계시.isin(range(6,23))]
```

```python
data_06_22.sort_values(by='집계시')
```

```python
data_06_22_counting = data_06_22['집계시'].value_counts()
```

```python
data_06_22_counting
```

```python
darta_06_22_counting_sorted = data_06_22_counting.sort_index()
```

```python
darta_06_22_counting_sorted
```

```python
plt.plot(darta_06_22_counting_sorted.index, darta_06_22_counting_sorted.values)
```

```python
x = darta_06_22_counting_sorted.index
```

```python
labels = [str(i) +'시' for i in x]
```

```python
labels
```

```python
values = darta_06_22_counting_sorted.values
```

```python
# configure figure size
plt.figure(figsize=(20,10))
# create line chart
plt.plot(labels, values, marker='s', color='r')
# Genrate labels and title
plt.title('집계시 기준 통행회수', fontsize=18)
plt.xlabel('집계시', fontdict={'size':16})
plt.ylabel('통행회수', fontdict={'size':16})
# show plot
plt.show()
```
