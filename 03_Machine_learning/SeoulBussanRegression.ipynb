#4. Linear Regression by Tensorflow

```python
import tensorflow as tf
import numpy as np
```

```python
Selected_Data = '2020-01-10' #@param {type:"date"}
input_data = int(Selected_Data.replace('-',''))
print(input_data)
```

```python
data_data = data[data['집계일자'] == input_data]
```

```python
data_data
```

```python
plt.plot(data_data['거리'], data_data['통행시간'], 'r*')
```

```python
data_time = data_data.groupby(['집계일자','거리'])['통행시간'].mean()
```

```python
data_time
```

```python
data_out = data_time.reset_index()
```

```python
data_out
```

```python
# Dataframe to List
data_list = data_out.values.tolist()
data_list[:3]
```

```python
x_train = [int(r[1]) for r in data_list]
y_train = [int(r[2]/10) for r in data_list]
```

```python
plt.plot(x_train, y_train, 'r*')
```

```python
learning_rate = 1e-8
learning_epochs = 2000
```

```python
# Stochastic gradient descent (SGD) Optimizer
sgd = tf.keras.optimizers.SGD(learning_rate=learning_rate)
# Mean Square Error (MSE) loss function
mse = tf.keras.losses.mean_squared_error

model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(1, input_dim=1))

model.compile(loss=mse, optimizer=sgd)
```

```python
model.summary()
```

```python
#Train the model
history = model.fit(x_train, y_train, epochs=learning_epochs)
```

```python
plt.plot(history.history['loss'])
plt.show()
```
